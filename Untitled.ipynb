```python
df['hour_sin'] = np.sin(2 * np.pi * df['hour'] / 24)
df['hour_cos'] = np.cos(2 * np.pi * df['hour'] / 24)
df['month_sin'] = np.sin(2 * np.pi * df['month'] / 12)
df['month_cos'] = np.cos(2 * np.pi * df['month'] / 12)
```

```python
def time_of_day(h):
    if 0 <= h < 6:
        return 'night'
    elif 6 <= h < 12:
        return 'morning'
    elif 12 <= h < 18: 
        return 'evening'
    else:
        return 'night'

df['time_of_day'] = df['hour'].apply(time_of_day)
```

```python
def haversine_km(lat1, lon1, lat2, lon2):
    R = 6371
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    dlat, dlon = lat2 - lat1, lon2 - lon1
    a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
    return 2 * R * np.arcsin(np.sqrt(a))

KTM_LAT, KTM_LON = 27.7172, 85.3240
df['dist_kathmandu_km'] = haversine_km(df['Latitude'], df['Longitude'], KTM_LAT, KTM_LON)

GORKHA_LAT, GORKHA_LON = 28.147, 84.708
df['dist_gorkha_km'] = haversine_km(df['Latitude'], df['Longitude'], GORKHA_LAT, GORKHA_LON)

# Coarse ~55km grid cells — groups nearby events without needing clustering yet
df['lat_bin'] = (df['Latitude'] // 0.5) * 0.5
df['lon_bin'] = (df['Longitude'] // 0.5) * 0.5
df['grid_cell'] = df['lat_bin'].astype(str) + '_' + df['lon_bin'].astype(str)
```

```python
def magnitude_category(m):
    if m < 4.0: return 'minor'
    elif m < 5.0: return 'light'
    elif m < 6.0: return 'moderate'
    elif m < 7.0: return 'strong'
    else: return 'major'
df['magnitude_category'] = df['Magnitude'].apply(magnitude_category)
```

```python
#the earthquake is divided into 3 bins shallow, intermediate and deep (depth from ground level<(70km)=shallow, d >= 70km && d < 300km = intermediate)
#nepal lies in shallow continental thrust system, which means earthquake occur from <= 30km so we are reformatting the bins for nepal
def depth_bin(d): 
    if d < 15: return 0 #shallow
    elif d < 35: return 1 #'intermediate'
    elif d < 70: return 2 #'strong'
    else: return 3 #'deep'

df["depth_bin"] = df['Depth_km'].apply(depth_bin)
```

```python
# Each whole-number jump in magnitude is roughly 32x more energy, not 2x —
# this keeps that nonlinearity available to the model
df['energy_joules'] = 10 ** (1.5 * df['Magnitude'] + 4.8)

# Simple heuristic, not a scientific formula: a strong, shallow quake is more
# dangerous than the same magnitude deep underground
df['magnitude_depth_ratio'] = df['Magnitude'] / (df['Depth_km'] + 1)
```

```python
#Spatial-Temporal Density
from sklearn.neighbors import BallTree

coords_rad = np.radians(df[['Latitude', 'Longitude']].values) #Converting coordinates to radian since geographical coordinates are in radian for haversine distance calculation
tree = BallTree(coords_rad, metric='haversine') #This index the neighbour within a radius
counts = tree.query_radius(coords_rad, r=50 / 6371, count_only=True) #expressed in 50km as radius since (50/6371 = 50km in radius since earth's radius = 6371km)
df['quakes_within_50km'] = counts - 1  # exclude the row itself(excluding the occured earthquake)

df = df.set_index('datetime')
df['quakes_last_30d'] = df['Magnitude'].rolling('30D').count() - 1 #looks back 30 days for each row of magnitude and counts how many earthwuake occured in the row(-1 refers excluding the occured earthquake itself)
df = df.reset_index() 

df['depth_bin'] = df['Depth_km'].apply(depth_bin)
df['energy_proxy'] = 10 ** (1.5 * df['Magnitude'])

# Confirm all 5 new columns exist before continuing
print(df[['dist_kathmandu_km', 'dist_gorkha_km',
          'quakes_within_50km', 'depth_bin', 'energy_proxy']].head())
```

```python
from sklearn.cluster import DBSCAN
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold, cross_val_predict
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
```

```python
#eps = eps is the neighbourhood radius that decides how far you look around each points
"""
Imagine we have points:
A, B, C, D, E, F plotted on a 2D plane.

Step 1: Pick eps (neighborhood radius)  
Suppose eps = 2 units. That means we draw a circle of radius 2 around each point to check which other points fall inside.

Step 2: Pick min_samples (minimum neighbors)  
Suppose min_samples = 3. That means a point must have at least 3 points (including itself) inside its eps circle to be considered a core point.

Point A: Within 2 units, it has neighbors B, C, and itself → total 3 points → core point.

Point B: Within 2 units, it has neighbors A, C, and itself → also a core point.

Point C: Same as above → core point.

Point D: Only has itself and maybe one neighbor → not enough → noise.

Point E: Close to A but doesn’t have enough neighbors itself → border point (still part of A’s cluster).

Point F: Far away, no neighbors → noise.

Since A, B, and C are all core points and connected (their neighborhoods overlap), they form Cluster 0.

E is a border point connected to A, so it also joins Cluster 0.

D and F don’t meet the density requirement → labeled as noise (-1).

So the labels might look like:
A=0, B=0, C=0, E=0, D=-1, F=-1
"""
```

```python
best_labels, best_n, best_eps = None, 0, None #Initializing variables to keep track of best clustering result
coords_full = np.radians(df[['Latitude', 'Longitude']].values) 
for eps_km in [20, 30, 40, 60, 80, 100, 130]: #eps = points that is taken to make the neighbours for eg., if eps = 20km radius then the points within 20km are considered to be neighbouts
    db = DBSCAN(eps=eps_km / 6371, min_samples= 3,
                metric='haversine')
    labels = db.fit_predict(coords_full) #assign each points to cluster label and if it doesn't fall on any cluster than it is considered as noise
    n_cl = len(set(labels)) - (1 if -1 in labels else 0) #counts number of clusters and if noise present the subtract by -1
    noise = (labels == -1).sum() #counts number of noise
    print(f"  eps={eps_km:3d} km → {n_cl:2d} clusters, {noise:3d} noise pts")
    if n_cl > best_n:
        best_n, best_labels, best_eps = n_cl, labels.copy(), eps_km 
df['cluster_raw'] = best_labels
print(f"\n  ✓ Best: eps={best_eps} km, {best_n} clusters")
```

```python
noise_mask = df['cluster_raw'] == -1 #creates a boolean mask for all rows which are labeled as noise
if noise_mask.sum() > 0:
    centroids = (df[df['cluster_raw'] != -1]  #this takes only the non-noise points, groups them by cluster label and computer the mean of latitude and longitude to give the centorid of clusters(center of mass of clusters)
                   .groupby('cluster_raw')[['Latitude', 'Longitude']]
                   .mean())
    ctree   = BallTree(np.radians(centroids.values), metric='haversine') #creates a spatial tree using the centroid of clusters
    _, idx  = ctree.query(np.radians(df.loc[noise_mask, 
                           ['Latitude', 'Longitude']].values), k=1) #for each noise points, queries the Balltree to find the nearest centroid and return closest clusterr(k=1)
    df.loc[noise_mask, 'cluster_raw'] = centroids.index[idx.flatten()]
```

```python
if best_n < 3:
    df['grid_score'] = (
        0.40 * (df['quakes_within_50km'] / max(df['quakes_within_50km'].max(), 1)) +
        0.35 * (df['Magnitude'] / df['Magnitude'].max()) +
        0.25 * (1 - df['depth_bin'] / 3)
    )
    q33, q66 = df['grid_score'].quantile([0.33, 0.66])
    df['risk_zone'] = pd.cut(df['grid_score'],
                              bins=[-np.inf, q33, q66, np.inf],
                              labels=['Low', 'Moderate', 'High'])
    df['risk_zone'] = df['risk_zone'].astype(str)
else:
       cstats = (df.groupby('cluster_raw')
                .agg(n_events=('Magnitude', 'count'),
                     mean_mag=('Magnitude', 'mean'),
                     max_mag=('Magnitude', 'max'))
                .copy())
```

```python
cstats['raw_score'] = (
        0.30 * cstats['n_events'] / cstats['n_events'].max() +
        0.35 * cstats['mean_mag'] / cstats['mean_mag'].max() +
        0.35 * cstats['max_mag']  / cstats['max_mag'].max()
)
 
q33, q66 = cstats['raw_score'].quantile([0.33, 0.66])
cstats['risk_zone'] = pd.cut(cstats['raw_score'],
                                   bins=[-np.inf, q33, q66, np.inf],
                                   labels=['Low', 'Moderate', 'High'])
cstats['risk_zone'] = cstats['risk_zone'].astype(str)
 
zone_map     = cstats['risk_zone'].to_dict()
df['risk_zone'] = df['cluster_raw'].map(zone_map)
 
print("  Cluster risk summary:")
print(cstats[['n_events', 'mean_mag', 'max_mag',
                   'raw_score', 'risk_zone']].sort_values('raw_score').to_string())
 
print()
print("  Final zone distribution:")
print(df['risk_zone'].value_counts().to_string())
print()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
FEATURES = [
    'Latitude',
    'Longitude', 
    'Depth_km',
    'dist_kathmandu_km',       
    'dist_gorkha_km', 
    'quakes_within_50km',
    'depth_bin',
    'energy_proxy'
]
X = df[FEATURES].values
le = LabelEncoder()
y = le.fit_transform(df['risk_zone'])
```

```python
rf = RandomForestClassifier(
    n_estimators=300,        # 300 trees — stable but not slow
    max_depth=8,             # prevents overfitting on 652 rows
    min_samples_leaf=3,      # each leaf needs at least 3 samples
    class_weight='balanced', # compensates for High > Moderate > Low counts
    random_state=42,
    n_jobs=-1                # use all CPU cores
)
```

```python
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
y_pred_cv = cross_val_predict(rf, X, y, cv=cv)
 
print("STEP 4 · Cross-validation results (5-fold stratified)")
print()
print(classification_report(y, y_pred_cv,
      target_names=le.classes_, digits=3))
print()
print("  Key metrics to focus on:")
print("  - F1-score: balanced accuracy for each zone")
print("  - Precision: of all 'High' predictions, how many were right?")
print("  - Recall   : of all actual 'High' zones, how many did we catch?")
print("  - For an alert system, HIGH RECALL for the High zone is critical.")
print()
```

Here from precision 0.994, almost all the predicted "High" cases were correct
Recall 0.947, captured most of the actual "High" cases, but missed few
F1 = 0.970, strong balance between precision and recall
Support = 566, This class had the largest number of sampples, so the model learned it very well

```python
def predict_risk_zone(lat, lon, depth_km, density_approx=None,
                      model_bundle=None):
    """
    Predict seismic risk zone for a new coordinate.
 
    Parameters
    ----------
    lat             : float  — latitude in decimal degrees
    lon             : float  — longitude in decimal degrees
    depth_km        : float  — focal depth in km
    density_approx  : int    — optional: estimated nearby quake count (default: dataset median)
    model_bundle    : dict   — loaded from risk_zone_model.pkl
 
    Returns
    -------
    zone  : str  — 'Low', 'Moderate', or 'High'
    probs : dict — {'High': 0.72, 'Low': 0.05, 'Moderate': 0.23}
    """
    if model_bundle is None:
        with open(MODEL_OUT_PATH, 'rb') as f:
            model_bundle = pickle.load(f)
 
    m_  = model_bundle['model']
    le_ = model_bundle['encoder']
 
    d_ktm   = haversine_km(lat, lon, KTM_LAT,   KTM_LON)
    d_gorka = haversine_km(lat, lon, GORKHA_LAT, GORKHA_LON)
    d_bin   = depth_bin(depth_km)
    energy  = 10 ** (1.5 * 5.0)   # baseline M5 energy for zone query
    density = density_approx if density_approx is not None else 10
 
    x_new   = np.array([[lat, lon, depth_km, d_ktm, d_gorka,
                          density, d_bin, energy]])
    proba   = m_.predict_proba(x_new)[0]
    label   = le_.inverse_transform([np.argmax(proba)])[0]
    probs   = {z: round(float(p), 3)
               for z, p in zip(le_.classes_, proba)}
    return label, probs
```
